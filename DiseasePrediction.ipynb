# <div class="alert alert-info">About Dataset</div>

#### Context
The dataset is the Cleveland Heart Disease dataset taken from the UCI repository. The dataset consists of 303 individuals’ data. There are 14 columns in the dataset(which have been extracted from a larger set of 75). No missing values. The classification task is to predict whether an individual is suffering from heart disease or not. (0: absence, 1: presence)

#### Content

The Cleveland Heart Disease database contains 13 attributes and a target variable. It has 8 nominal values and 5 numeric values. The detailed description of all these features are as follows:

1. Age: Patients Age in years (Numeric)<br>
2. Sex: Gender (Nominal)<br>
    0 = Female <br>
    1 = Male <br>
3. cp: Type of chest pain experienced by patient. This term categorized into 4 category.<br>
    0 = typical angina <br>
    1 = atypical angina <br>
    2 = non- anginal pain <br>
    3 = asymptomatic (Nominal)
4. trestbps: patient's level of blood pressure at resting mode in mm/HG (Numerical)
5. chol: Serum cholesterol in mg/dl (Numeric)
6. fbs: Blood sugar levels on fasting > 120 mg/dl represents as 1 in case of true and 0 as false (Nominal)
7. restecg: Result of electrocardiogram while at rest are represented in 3 distinct values<br>
    0 = Normal <br>
    1 = having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of >0.05 mV)  <br>
    2 = showing probable or definite left ventricular hypertrophyby Estes' criteria (Nominal)
8. thalach: Maximum heart rate achieved (Numeric)
9. exang: Angina induced by exercise (Nominal)<br>
    0 = NO <br>
    1 = Yes 
10. oldpeak: Exercise induced ST-depression in relative with the state of rest (Numeric)
11. slope: ST segment measured in terms of slope during peak exercise<br>
    0 = up sloping; <br>
    1 = flat; <br>
    2 = down sloping(Nominal) 
12. ca: The number of major vessels colored by fluoroscopy (0–3)(nominal)
13. thal: A blood disorder called thalassemia<br>
    0 = NULL 
    1 = normal blood flow 
    2 = fixed defect (no blood flow in some part of the heart) 
    3 = reversible defect (a blood flow is observed but it is not normal(nominal)
14. target: It is the target variable which we have to predict 1 means patient is suffering from heart disease and 0 means patient is normal.

# <div class="alert alert-info">Module 4 - Preparing the dataset</div>

## Encoding Categorical Features

We implement one-hot encoding to convert categorical features in our dataset, namely 'cp,' 'thal,' and 'slope,' into a numerical format. By executing this code, we successfully transform these categorical variables, facilitating their integration into our analysis. This encoding plays a pivotal role in our project, enabling us to seamlessly work with categorical data and uncover valuable insights into the influence of these features on health conditions and outcomes. The resultant 'df_encoded' dataset is now primed for advanced analysis and model building.

```python
categorical_cols = ['cp', 'thal','slope']
df[categorical_cols] = df[categorical_cols].astype(int)

df_encoded = pd.get_dummies(df,
                            columns=categorical_cols,
                            prefix_sep='_',
                            dtype=int)
df_encoded = df_encoded.astype(int)

#--- Inspect data ---
df_encoded.dtypes
```

## Preparing Features and Target Variable

For the subsequent phase, we prepare the features and the target variable for our analysis. We establish the variable 'x' by excluding the 'condition' column, forming our feature set. Simultaneously, the 'y' variable is defined to encompass the 'condition' column, signifying our target variable. This demarcation is pivotal for our project, laying the foundation for extensive data analysis, modeling, and comprehending the intricate relationship between the features and health condition outcomes within our dataset.

```python
x = df_encoded.drop('condition', axis=1)
y = df_encoded['condition']
```

## Scaling Features

We use the MinMaxScaler from the sklearn library to scale the feature set 'x.' Scaling is crucial for ensuring that all the features are on a similar scale, preventing any feature from dominating the analysis due to its magnitude. By applying the MinMaxScaler, we transform the features to a common range, which is vital for the accuracy and effectiveness of our analysis and modeling in our project.



```python
from sklearn.preprocessing import MinMaxScaler

# creating a Scaler object
minmax = MinMaxScaler()
x = minmax.fit_transform(x)

#--- Inspect data ---
x
```

## Splitting the Data into Training and Testing Sets

As the final step in data preparation, we split our dataset into training and testing sets using the train_test_split function from the sklearn library. The training set, 'X_train' and 'Y_train,' is designed to train our predictive models, while the testing set, 'X_test' and 'Y_test,' is reserved for evaluating the model's performance. By performing this data split, we ensure that our models are trained and tested on different data subsets, which is essential for assessing their accuracy and generalization to new data in our project.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test  = train_test_split(x,
                                                    y,
                                                    train_size = 0.8,
                                                    test_size=0.2,
                                                    random_state = 4)
```

# <div class="alert alert-info">Module 5 - Model Building</div>

## Model 1 - Logistic Regression Model

A logistic regression model has been constructed for our project utilizing the sklearn library. The 'lr_model' is created and trained with the provided training data, 'X_train' and 'Y_train.' Furthermore, we evaluate the model's performance through cross-validation with 10 folds, estimating its accuracy. The 'lr_mean_score' signifies the mean accuracy across the folds.

Upon analysis, we observe that, for our dataset, the logistic regression model, when subjected to 10-fold cross-validation, yields a mean accuracy of 0.84 with a standard deviation of 0.07.


```python
#--- Import LogisticRegression ---
from sklearn.linear_model import LogisticRegression

#--- Import cross_val_score from sklearn.model_selection ---
from sklearn.model_selection import cross_val_score

# Create a model
lr_model = LogisticRegression()

# Train the model
lr_mod = lr_model.fit(X_train, Y_train)
 
# Perform 10 folds cross-validation
lr_cv_results = cross_val_score(lr_model, X_train, Y_train, cv=10)
lr_mean_score  = lr_cv_results.mean()

#--- model performance ---
print(f"Model score: {lr_mean_score:.2f} accuracy with a standard deviation of {lr_cv_results.std():.2f}")
```
